# Concatenating the Records DataFrames

```python
records_df = pd.DataFrame()
records_df = pd.concat([year_2015_df, year_2016_df, year_2017_df, year_2018_df, year_2019_df, year_2020_df, year_2021_df, 
                       year_2022_df, year_2023_df, year_2024_df, year_2025_df])
records_df.head()
```

# Merging team_batting_df and records_df

```python
df = pd.merge(team_batting_df,
              records_df,
              how='inner',
              on='index'
)
df.tail()
```

```python
df['W'].describe()
```

```python
df['Wins_Grouped'] = pd.cut(
                            df['W'],
                            bins = [0, 68.1, 80.1, 89.1, 112],
                            labels = ['First Quartile', 'Second Quartile', 'Thrid Quartile', 'Fourth Quartile'],
                            include_lowest = True)
df.head()
```

```python
imp_stats = df[['Team', 'AB', 'AVG', 'OBP', 'OPS', 'HR', 'RBI', 'K%', 'Hard%', 'SB']]
imp_stats.head()
```

```python
sb_percentage = df['SB'] / (df['SB'] + df['CS'])
imp_stats['SB%'] = sb_percentage
imp_stats.head()
```

```python
imp_stats['Wins_Grouped'] = df['Wins_Grouped']
imp_stats.head()
```

```python
groupby_wins = imp_stats.groupby('Wins_Grouped').describe()
groupby_wins = groupby_wins.T
groupby_wins
```

```python
corr_df = imp_stats
corr_df = corr_df.drop(['Wins_Grouped', 'Team', 'AB'], axis=1)
corr_df['Wins'] = df['W']
print(corr_df.corr())
```
